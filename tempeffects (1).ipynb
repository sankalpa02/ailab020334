```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the probability of accepting an inferior node (worse solution)
def acceptance_probability(delta_energy, temperature):
    return np.exp(-delta_energy / temperature)

# Function to simulate the effect of temperature on the acceptance of inferior solutions
def simulate_temperature_effect(initial_temperature, cooling_rate, max_iterations):
    temperatures = []
    probabilities = []

    # Simulate the process for a range of temperatures
    for t in np.linspace(initial_temperature, 0.1, max_iterations):
        # Randomly generate a delta energy (difference between current and new solution)
        delta_energy = np.random.uniform(0, 10)  # Assume worst solution could be 10 units worse

        # Calculate the probability of accepting the new solution based on the current temperature
        prob = acceptance_probability(delta_energy, t)

        # Store values for plotting
        temperatures.append(t)
        probabilities.append(prob)

    return temperatures, probabilities

# Main function to run the simulation and plot the results
def main():
    initial_temperature = 100  # Starting temperature
    cooling_rate = 0.99  # Cooling rate, not actually used in the simulation here, but important in actual annealing
    max_iterations = 100  # Number of iterations to simulate

    temperatures, probabilities = simulate_temperature_effect(initial_temperature, cooling_rate, max_iterations)

    # Plot the results
    plt.figure(figsize=(8, 6))
    plt.plot(temperatures, probabilities, label='Probability of Accepting Inferior Node', color='b')
    plt.xlabel('Temperature')
    plt.ylabel('Probability of Accepting Inferior Node')
    plt.title('Effect of Temperature on Acceptance Probability in Simulated Annealing')
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
